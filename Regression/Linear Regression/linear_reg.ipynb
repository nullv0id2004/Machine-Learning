```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df
```

```python
for column in df.columns:
    df[column] = (df[column]-df[column].mean())/df[column].std()
df
    
```

```python
df.corr()['SalePrice']
```

```python
df_np = df.to_numpy()
df_np.shape
```

```python
X_train , Y_train = df_np[:,:10] , df_np[:,-1]
X_train.shape , Y_train.shape
```

```python
X_train
```

```python
test_df = pd.read_csv('test.csv')

test_df

```

```python
for column in test_df.columns:
    test_df[column] = (test_df[column]-test_df[column].mean())/test_df[column].std()
test_np = test_df.to_numpy()
test_df
```

```python
X_test = test_np[:,:10]
Y_test = test_np[:,-1]
X_test , Y_test.shape
```

```python
def get_predictions(model,X):
    (n,p_minus_one) = X.shape
    p = p_minus_one +1

    new_X = np.ones(shape=(n,p))
    new_X[:,1:] = X

    return np.dot(new_X,model)
```

```python
from numpy.linalg import inv

def get_best_model(X, y):

  (n, p_minus_one) = X.shape
  p = p_minus_one + 1

  new_X = np.ones(shape=(n, p))
  new_X[:, 1:] = X

  return np.dot(np.dot(inv(np.dot(new_X.T, new_X)), new_X.T), y)
```

```python
best_model = get_best_model(X_train, Y_train)
test_df['Predicted Sale Price'] = get_predictions(best_model, X_test)
test_df
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mean_absolute_error(test_df['Predicted Sale Price'], Y_test), mean_squared_error(test_df['Predicted Sale Price'], Y_test), r2_score(test_df['Predicted Sale Price'],Y_test)
```
