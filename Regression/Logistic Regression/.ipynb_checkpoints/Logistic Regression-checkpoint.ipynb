```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df = df.drop(columns=['Name'])
df
```

```python
df['Sex'].unique()
df = pd.get_dummies(df, columns=['Sex'])
df = df.replace({True: 1, False: 0})
df
```

```python
df2= df.loc[:,'Pclass':'Sex_male']
df2
```

```python
for column in df2.columns:
    df2[column] = (df2[column]-df2[column].mean())/df2[column].std()
df2
```

```python
df.corr()['Survived']
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
Dict = {'lr':0.1, 'n_iters':10000, 'weights':None, 'bias':None}
```

```python
df_np = df.to_numpy()
df2_np = df2.to_numpy()
df_np.shape , df2_np.shape
```

```python
X_train ,Y_train = df2_np, df_np[:,-8]
X_train , Y_train
```

```python
def fit(Dict, X, y):
    n_samples, n_features = X.shape
    Dict['weights'] = np.zeros(n_features)
    Dict['bias'] = 0
    for _ in range(Dict['n_iters']):
        linear_pred = np.dot(X, Dict['weights']) + Dict['bias']
        predictions = sigmoid(linear_pred)
        

        dw = (1/n_samples) * np.dot(X.T, (predictions - y))
        db = (1/n_samples) * np.sum(predictions-y)

        Dict['weights'] = Dict['weights'] - Dict['lr']*dw
        Dict['bias'] = Dict['bias'] - Dict['lr']*db

```

```python
def predict(Dict, X):
    linear_pred = np.dot(X, Dict['weights']) + Dict['bias']
    y_pred = sigmoid(linear_pred)
    class_pred = [0 if y<=0.5 else 1 for y in y_pred]
    return class_pred
```

```python
test_df = pd.read_csv('test.csv')
test_df
```

```python
test_df = test_df.drop(columns=['Name'])
test_df
```

```python
test_df['Sex'].unique()
test_df = pd.get_dummies(test_df, columns=['Sex'])
test_df = test_df.replace({True: 1, False: 0})
test_df
```

```python
test_df2 = test_df.loc[:,'Pclass':'Sex_male']
test_df2
```

```python
for column in test_df2.columns:
    test_df2[column] = (test_df2[column]-test_df2[column].mean())/test_df2[column].std()
test_df2
```

```python
test_np = test_df.to_numpy()
test2_np = test_df2.to_numpy()
test_np.shape , test2_np.shape
```

```python
X_test ,Y_test = test2_np, test_np[:,-8]
X_test , Y_test
```

```python
fit(Dict,X_train,Y_train)
y_pred = predict(Dict,X_test)
y_pred = np.array(y_pred)
test_df['Survival Prediction'] = y_pred
test_df
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_df['Survival Prediction'],Y_test)
```
